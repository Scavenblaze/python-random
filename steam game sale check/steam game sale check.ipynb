```python
from bs4 import BeautifulSoup
import requests
import lxml
```

```python
#edit url with the desired steam game's url
url = "https://store.steampowered.com/app/815370/Green_Hell/"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'lxml')
soup
```

```python
a = soup.title.get_text().split()
a
```

```python
for text in a:
    if text == "Save":
        print(f"The game is on sale with a discount of {a[1]}")
        break
    else: print("game is not on sale")
```
